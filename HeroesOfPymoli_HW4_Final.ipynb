## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#extracting raw data 
Player_Buyers = pd.DataFrame(purchase_data[{"Age","Price"}])

#--------------------------------------------------------------------------------------

#Calculating players per age range
Age_count = pd.DataFrame(purchase_data[{"Age","SN"}])
Age_count = Age_count.drop_duplicates("SN", inplace = False)
bins_count = [0, 9, 14, 19, 24, 29, 34, 39, 40]
ranges_count = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
pd.cut(Age_count["Age"], bins_count, labels=ranges_count).head()
Age_count["Age Frame"] = pd.cut(Age_count["Age"], bins_count, labels=ranges_count)
Age_count = Age_count.groupby("Age Frame")
Age_frame = Age_count.count()

#------------------------------------------------------------------------------------------------

# structure table: create bins and an additional column added to help calculate the aver purchase price per person 
bins_age = [0, 9, 14, 19, 24, 29, 34, 39, 40]
ranges_age = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
pd.cut(Player_Buyers["Age"], bins_age, labels=ranges_age).head()
Player_Buyers["Range Age"] = pd.cut(Player_Buyers["Age"], bins_age, labels=ranges_age)

Age_Metrics = Player_Buyers.groupby(["Range Age"])
Age_Metrics = Age_Metrics.agg({"Age":["count"],"Price":["mean"]})
Age_Metrics["Total Purchase Value"] = Age_Metrics.iloc[:,0] * Age_Metrics.iloc[:,1]
Age_Metrics["Total Members of Age Range"] = Age_frame["SN"]
Age_Metrics.loc[:,"Avg Total Purchase per Person"] = Age_Metrics.iloc[:,2] / Age_Metrics["Total Members of Age Range"]
Age_Metrics = Age_Metrics.rename(columns={"Age":"Purchase"})
del Age_Metrics["Total Members of Age Range"]

Age_Metrics
```

```python
#Raw Data table
SN_RawData = pd.DataFrame(purchase_data[{"SN","Item Name","Price"}])
SN_Metrics = SN_RawData.groupby(["SN"])

#Create table
SN_Metrics = SN_Metrics.agg({"Item Name":["count"],"Price":["mean"]})
SN_Metrics["Total Purchase Value"] = SN_Metrics.iloc[:,0] * SN_Metrics.iloc[:,1]
SN_Metrics = SN_Metrics.rename(columns={"Item Name":"Total Purchase"})

#Sort 
SN_Metrics = SN_Metrics.sort_values("Total Purchase Value", ascending=False)
SN_Metrics.head()

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Raw Data
RawData = pd.DataFrame(purchase_data[{"Item ID","Item Name","Purchase ID","Price"}])
Items_Metrics = RawData.groupby(["Item ID","Item Name"])

# construct table, price mean = actual price b/c it's constant per item
Items_Metrics = Items_Metrics.agg({"Purchase ID":["count"],"Price":["mean"]})
Items_Metrics["Total Purchase Count"] = Items_Metrics.iloc[:,0] * 1
Items_Metrics["Total Purchase Value"] = Items_Metrics.iloc[:,0] * Items_Metrics.iloc[:,1]
Items_Metrics = Items_Metrics.sort_values("Total Purchase Count", ascending=False)
del Items_Metrics['Purchase ID']
Items_Metrics.head()

```
